```python
import folium

# Create a map centered around the Gulf of Mexico
m = folium.Map(location=[27.0, -89.5], zoom_start=6)

# Data for the projects with accurate locations in the Gulf of Mexico
projects = [
    {'name': 'Beacon Zephyrus', 'location': [28.0, -89.5], 'details': 'Beacon Zephyrus at Mississippi Canyon Block 759, water depth: 3,530 feet'},
    {'name': 'Shell Vito WF', 'location': [28.0, -89.2], 'details': 'Shell Vito WF at Mississippi Canyon, water depth: 4,000 feet'},
    {'name': 'Dover', 'location': [28.1, -89.3], 'details': 'Dover at Mississippi Canyon Block 612, water depth: 2,286 m'},
    {'name': 'Sparta', 'location': [27.5, -92.0], 'details': 'Sparta in the Garden Banks area, 120 miles from southwest Louisiana’s coast'}
]

# Add markers for each project
for project in projects:
    folium.Marker(
        location=project['location'],
        popup=f"{project['name']}: {project['details']}",
        tooltip=project['name']
    ).add_to(m)

# Save the map to an HTML file
m.save('gulf_of_mexico_projects_map.html')

# Display map in a Jupyter notebook (if applicable)
m
```
